```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
snsblue = sns.color_palette('pastel')[0]
df = pd.read_csv('gutenberg_readability.csv', index_col=0)
md = pd.read_csv('metadata-pg-genres-subset.tsv', sep='\t', index_col='id')
```

```python
md
```

```python
subset = df.index.intersection(md.index)
df = df.loc[subset, :]
md = md.loc[subset, :]

genres = ['Western', 'Adventure', 'Love', 'War', 'Humour', 'Sea',
                'Pulp fiction', 'SciFi & Fantasy', 'General fiction',
                'Mystery & Detective stories', 'Juvenile', 'Translations']
```

```python
fig, axes = plt.subplots(5, 2, figsize=(10, 8), sharey=False, sharex=True)
column = 'ARI'
for ax, genre in zip(axes.flatten(), genres):
    df.loc[md[genre] == 1, column].hist(ax=ax)
    ax.set_title(f'{genre}'); ax.set_ylabel('# texts'); ax.set_xlabel(column)
fig.tight_layout()
```

```python
column = 'type_token_ratio'
order = pd.Series({
        genre: df.loc[md[genre] == 1, column].mean() for genre in genres}).sort_values()
ax = sns.violinplot(
        data=[df.loc[md[genre] == 1, column] for genre in order.index],
        orient='horiz');
ax.set_yticklabels(order.index);
```

```python
column = 'words_per_sentence'
order = pd.Series({
        genre: df.loc[md[genre] == 1, column].mean() for genre in genres}).sort_values()
ax = sns.violinplot(
        data=[df.loc[md[genre] == 1, column] for genre in order.index],
        orient='horiz');
ax.set_yticklabels(order.index);
```

```python
columns = ['DaleChallIndex', 'GunningFogIndex', 'FleschReadingEase', ]
fig, axes = plt.subplots(3, 1, figsize=(8, 11))
for column, ax in zip(columns, axes):
    # FIXME: should order be the same for all plots, to facilitate comparison?
    # or use same color for each genre
    order = pd.Series({
        genre: df.loc[md[genre] == 1, column].mean() for genre in genres}).sort_values(
            ascending=True if column == 'FleschReadingEase' else False)
    sns.violinplot(
            data=[df.loc[md[genre] == 1, column] for genre in order.index],
            orient='horiz', color=snsblue, ax=ax);
    ax.set_yticklabels(order.index);
    ax.set_xlabel(column);
fig.tight_layout(); fig.savefig('fig/fig5.12readabilityviolinplot.pdf'); fig.savefig('fig/fig5.12readabilityviolinplot.png', dpi=150); 
```

```python
columns = ['DaleChallIndex', 'GunningFogIndex', 'FleschReadingEase', ]
fig, axes = plt.subplots(3, 1, figsize=(8, 11))
for column, ax in zip(columns, axes):
    # FIXME: should order be the same for all plots, to facilitate comparison?
    order = pd.Series({
        genre: df.loc[md[genre] == 1, column].mean() for genre in genres}).sort_values(
            ascending=True if column == 'FleschReadingEase' else False)
    sns.boxplot(
            data=[df.loc[md[genre] == 1, column] for genre in order.index],
            orient='horiz', color=snsblue, ax=ax);
    ax.set_yticklabels(order.index);
    ax.set_xlabel(column);
# fig.tight_layout(); fig.savefig('fig/readabilityboxplot.pdf'); fig.savefig('fig/readabilityboxplot.png', dpi=150); 
```

```python
# Create a new version of the dataframe with one row for each instance of (ID, genre_label).
# This means that texts with multiple genre labels are represented as multiple rows in this dataframe!
df1 = pd.concat([md, df], axis=1)
concatted = pd.concat([df1.loc[df1[genre] == 1, :] for genre in genres], axis=0)
concatted.insert(
        0, 'genre',
        [genre
            for genre in genres
                for instance in range(df1[genre].sum())])
concatted
```

```python
fig, ax = plt.subplots(figsize=(8, 4))
ax = sns.violinplot(data=concatted, x='directspeech_ratio', y='genre', color=snsblue,
               order=concatted.groupby('genre')['directspeech_ratio'].median().sort_values().index);
ax.set_ylabel(None); ax.figure.tight_layout();
ax.figure.savefig('fig/fig5.13directspeech_ratio.pdf'); ax.figure.savefig('fig/fig5.13directspeech_ratio.png', dpi=150)
```
